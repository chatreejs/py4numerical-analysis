## 1. Random Search

* A brute-force method
    * Based on evaluation of the function randomly at selected values of the independent variables.
    * If a sufficient number of samples are conducted, the optimun will be eventually located.

* Advantages:
    * Works even for discontinuous and non-differentiable functions
    * Always finds the global optimun rather than the global minimum

* Disadvantages:
    * As the number of independent variables grows, the task can become onerous.
    * Not efficient, it does not account for the behavior of underlying function.

```python
def brute_force(f, start, end):
    try:
        if start > end :
            raise ValueError('start must be <= end')
            
        minimum = float('inf')
        x = start
        
        search_range = np.arange(start, end, 0.0001)
        for i in search_range:
            y = f(i)
            if y < minimum:
                minimum = y
                x = i
        return (x, minimum)
            
    except ValueError as e :
        print('ValueError Exception : ', e)
        traceback.print_exception()
```

### Example
find a minimum of $f(x) = \frac{x^2}{10} - 2 \sin{x}$ on the interval $[-8 , 8]$

```python
f = lambda x : (np.power(x, 2)) / 10 - (2 * np.sin(x))
minimum = brute_force(f, -8, 8)
print("minimum (x, y): (%.4f, %.4f)" % (minimum[0], minimum[1]))
```

### Visualization

```python
xs = np.linspace(-10, 10, 100)
ys = (np.power(xs, 2)) / 10 - (2 * np.sin(xs))

plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.scatter(1.4276, -1.7757, c='k', s=6)
plt.plot(xs, ys, 'r')
plt.show()
```
